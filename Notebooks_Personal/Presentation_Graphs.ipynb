```python
import pandas as pd
import seaborn as sns


df_train = pd.read_csv("../data/zindi/train.csv")
sns.set_style("whitegrid")
```

### Aircaft Codes

```python
from matplotlib import pyplot as plt

df_train['AC'].value_counts().plot(kind='bar', figsize=(10,5))
plt.xlabel('Aircraft Code')
plt.ylabel('Count')
plt.title('Distribution of Aircraft Codes\n', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
mean_value = (
    df_train.groupby('AC')
      .agg(
          mean_value=('target', 'mean'),
          count=('target', 'size')
      )
      .sort_values('count', ascending=False)['mean_value']
)

mean_value.plot(kind='bar', figsize=(10,5))
plt.xlabel('Aircraft Code')
plt.ylabel('Mean Delay')
plt.title('Mean Delay per Aircraft Code\n', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

prob_over_0 = (
    df_train.assign(over_0=df_train['target'] > 0)
      .groupby('AC')
      .agg(
          prob_over_0=('over_0', 'mean'),
          count=('over_0', 'size')
      )
      .sort_values('count', ascending=False)['prob_over_0']
)

prob_over_0.plot(kind='bar', figsize=(10,5))
plt.xlabel('Aircraft Code')
plt.ylabel('P(Value > 0)')
plt.title('Probability of Delay per Aircraft Code\n', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Flight Duration

```python
from Utils.Transformers import TypeCastDatetimeTransformer, DatetimeDifferenceTransformer
from sklearn.pipeline import Pipeline

flight_duration_pipeline =  Pipeline([
    ("scheduled_dep_time_to_daytime", TypeCastDatetimeTransformer("STD")),
    ("scheduled_arr_time_to_daytime", TypeCastDatetimeTransformer("STA", "%Y-%m-%d %H.%M.%S")),

    ("compute_flight_time", DatetimeDifferenceTransformer("STD", "STA", "flight_duration_hours", "hours")),
])

df_train = flight_duration_pipeline.fit_transform(df_train)
```

```python
plt.figure(figsize=(10,5))
plt.hist(df_train['flight_duration_hours'], bins=30)
plt.xlabel('Flight Duration (Hours)')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Duration (Hours) before Capping\n', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
df_train = flight_duration_pipeline.fit_transform(df_train)
df_train["flight_duration_hours"] = df_train['flight_duration_hours'].clip(upper=12)

plt.figure(figsize=(10,5))
plt.hist(df_train['flight_duration_hours'], bins=30)
plt.xlabel('Flight Duration (Hours)')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Duration (Hours) after Capping at 12 Hours\n', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
